# Quantum Noise Simulation

This notebook demonstrates quantum noise simulation for Thrust 1.

## Goals:
1. Simulate quantum circuits with noise
2. Characterize noise effects

```python
# Import packages
import sys
sys.path.append('../src')

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram
import pandas as pd

# Set style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")

print("Packages imported successfully!")
print(f"NumPy: {np.__version__}")
print(f"Qiskit: {import qiskit; qiskit.__version__}")
```

```python
# Create a simple quantum circuit
def create_bell_state():
    """Create a Bell state circuit."""
    qc = QuantumCircuit(2, 2)
    qc.h(0)
    qc.cx(0, 1)
    qc.measure([0, 1], [0, 1])
    return qc

# Test the circuit
qc = create_bell_state()
print("Circuit:")
print(qc)
print(f"\nNumber of qubits: {qc.num_qubits}")
print(f"Circuit depth: {qc.depth()}")
```

```python
# Simulate the circuit
backend = Aer.get_backend('qasm_simulator')
result = execute(qc, backend, shots=1024).result()
counts = result.get_counts(qc)

print("Measurement counts:")
for state, count in counts.items():
    print(f"  {state}: {count} ({count/1024*100:.1f}%)")

# Plot results
fig = plot_histogram(counts)
fig.suptitle('Bell State Measurement Results', fontsize=14)
plt.show()
```
